```python
import pandas as pd
```

```python
# Load CSV file with semicolon separator
df02 = pd.read_csv("msn_02_fuel_leak_signals_preprocessed.csv", sep=";")
```

# Prepping the dataset 

```python
# SETTING INDEX AS A DATETIME
df02['UTC_TIME'] = pd.to_datetime(df02['UTC_TIME'])
df02 = df02.set_index("UTC_TIME")

# Drop time, month, day, and year column because they are already in the UTC Time im assuming
df02 = df02.drop(columns=['time', 'month', 'day', 'year'])
```

```python
df02 = df02.drop_duplicates()
```

```python
df02.info(verbose=True, show_counts=True)
```

```python
# Check missing values percentage
missing_values = (df02.isnull().sum() / len(df02)) * 100
print(missing_values[missing_values > 0].sort_values(ascending=False))
```

### Dealing with missing value

```python

# Dropping columns with more then 50% missing values
threshold = 50
columns_to_drop = missing_values[missing_values > threshold].index
df02 = df02.drop(columns=columns_to_drop)

# For numerical columns, impute using the median
numerical_cols = df02.select_dtypes(include=['float64', 'int64']).columns
df02[numerical_cols] = df02[numerical_cols].fillna(df02[numerical_cols].median())

# For categorical columns, impute using the mode (most frequent value)
categorical_cols = df02.select_dtypes(include=['object']).columns
df02[categorical_cols] = df02[categorical_cols].fillna(df02[categorical_cols].mode().iloc[0])
```

```python
import matplotlib.pyplot as plt
# Plot boxplots
plt.figure(figsize=(15, 8))
df02[numerical_features_02].boxplot(rot=45)
plt.title("Boxplots of Numerical Features in MSN 02")
plt.show()
```

```python


#going to ignore the ones that are booleans for the box plot


categories = {
    'Fuel Flow': [
        'FUEL_FLOW_1', 'FUEL_FLOW_2', 'FUEL_FLOW_3', 'FUEL_FLOW_4'
    ],
    'Fuel Used': [
        'FUEL_USED_1', 'FUEL_USED_2', 'FUEL_USED_3', 'FUEL_USED_4'
    ],
    'Fuel Leak Detection': [
        'LEAK_DETECTION_LEAK_FLOW'
    ],
    'Fuel Quantity': [
        'VALUE_FUEL_QTY_CC1', 'VALUE_FUEL_QTY_CC2', 'VALUE_FUEL_QTY_CC3', 'VALUE_FUEL_QTY_CC4',
        'VALUE_FUEL_QTY_CT', 'VALUE_FUEL_QTY_FT1', 'VALUE_FUEL_QTY_FT2', 'VALUE_FUEL_QTY_FT3',
        'VALUE_FUEL_QTY_FT4', 'VALUE_FUEL_QTY_LXT', 'VALUE_FUEL_QTY_RXT'
    ],
    'Fuel Volume': [
        'VALUE_FUEL_VOL_LST', 'VALUE_FUEL_VOL_RST'
    ],
    'Altitude and Position': [
        'FW_GEO_ALTITUDE', 'SELECTED_GADIR_ALTITUDE_VALUE'
    ],
    'Flight Status': [
        'FLIGHT_PHASE_COUNT', 'PITCH_ANGLE', 'ROLL_ANGLE'
    ]
}
```

```python
for category, columns in categories.items():
    # Ensure all columns exist in the DataFrame
    columns_in_df = [col for col in columns if col in df02.columns]
    
    # Plot boxplot if there are valid columns
    if columns_in_df:
        plt.figure(figsize=(10, 6))
        df02[columns_in_df].boxplot(rot=45)
        plt.title(f"Boxplot of {category}")
        plt.show()
```

```python
# Set up the figure with subplots (using len(categories) for the number of rows)
fig, axes = plt.subplots(len(categories), 1, figsize=(12, 4 * len(categories)))

# If there's only one category, axes will not be an array, so we handle that case
if len(categories) == 1:
    axes = [axes]

# Plot histograms for each category
for idx, (category, columns) in enumerate(categories.items()):
    ax = axes[idx]
    # Plot each histogram for the category's columns
    ax.hist(df02[columns].values.flatten(), bins=50, alpha=0.6, color='blue')
    ax.set_title(f'Histogram for {category}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Set up the figure with subplots (using len(categories) for the number of rows)
fig, axes = plt.subplots(len(categories), 1, figsize=(12, 4 * len(categories)))

# If there's only one category, axes will not be an array, so we handle that case
if len(categories) == 1:
    axes = [axes]

# Colors to use for each column
colors = ['blue', 'green', 'orange', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

# Plot histograms for each category
for idx, (category, columns) in enumerate(categories.items()):
    ax = axes[idx]
    # Plot each histogram for the category's columns with a different color
    for i, column in enumerate(columns):
        ax.hist(df02[column].dropna(), bins=50, alpha=0.6, color=colors[i % len(colors)], label=column)

    ax.set_title(f'Histogram for {category}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.legend()  # Show legend with column names

# Adjust layout
plt.tight_layout()
plt.show()
```
